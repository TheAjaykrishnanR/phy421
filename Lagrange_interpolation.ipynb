```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x*np.pi/6)
```

```python
x_pts = np.array([-1, 1, 3, 5])
y_pts = f(x_pts)

x = np.arange(0, 10, 0.1)

plt.plot(x, f(x))
```

```python
# 

def Lagrange(x: np.array, x_pts, y_pts):
    n = len(x_pts)

    def L_k(x, i):
        prod = 1
        for j in range(n):
            if i != j:
                prod = prod*(x-x_pts[j])/(x_pts[i] - x_pts[j])
        return prod

    out = 0
    for i in range(n):
        out += y_pts[i]*L_k(x, i)
    return out
```

```python
inter_y = Lagrange(x, x_pts, y_pts)
plt.plot(x, f(x))
plt.plot(x, inter_y)
plt.show()
```
